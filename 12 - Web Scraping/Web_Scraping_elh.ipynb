# Mars Hemispheres

```python
Hemisphere_url = "https://marshemispheres.com/"
browser.visit(Hemisphere_url)
```

```python
html = browser.html
soup = BeautifulSoup(html, "lxml")
```

```python
Hemisphere = soup.find_all("div", {"class":"item"})
Hemisphere
```

```python
Hemisphere_info = []
for Hemisphere in Hemisphere:
    Hemisphere_url_final = Hemisphere_url + Hemisphere.find("a", {"class": "itemLink"})["href"]
    browser.visit(Hemisphere_url_final)
    time.sleep(1)
    
    html = browser.html
    soup = BeautifulSoup(html, "lxml")
    
    link = Hemisphere_url + soup.find("img", {"class", "wide-image"})["src"]
    title = soup.find("h2", {"class", "title"}).text
    
    data = {
        "img_url": link,
        "title": title
    }
    
    Hemisphere_info.append(data)
```
